```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import products from json an run basic analysis on it to see how labels are distributed

```python
products = pd.read_json('../data/combined.json', orient='records')
products.set_index('id', inplace=True)
products.head()
```

```python
product_counts = products['category'].value_counts()
plt.bar(product_counts.index, product_counts)
plt.xticks(rotation=45)
```
